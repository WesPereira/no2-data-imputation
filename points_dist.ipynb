```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from shapely.geometry import Point

# Converter o GeoJSON em um DataFrame GeoPandas
gdf = gpd.GeoDataFrame.from_features(geojson['features'])
gdf['geometry'] = gdf['geometry'].apply(Point)

# Carregar o shapefile do estado do Pará (ou outro estado desejado)
# Certifique-se de ter um shapefile do estado do Pará ou ajuste o caminho abaixo
# Aqui estamos usando um exemplo, então ajuste conforme necessário
estado_para = gpd.read_file('PA_UF_2022/PA_UF_2022.shp')

# Definir uma projeção para o GeoDataFrame (usando a projeção WGS84)
gdf.crs = "EPSG:4326"

# Definir limites para mostrar mais do Brasil
ax = estado_para.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
ax.set_xlim([-75, -30])  # Ajuste conforme necessário
ax.set_ylim([-15, 5])    # Ajuste conforme necessário

# Adicionar o fundo do mapa da Contextily (OpenStreetMap neste caso)
ctx.add_basemap(ax, crs=gdf.crs)

# Plotar os pontos em cima do estado
gdf.plot(ax=ax, color='darkgreen', edgecolor='black', markersize=50, label='pontos de coleta')

# Adicionar título
plt.title('Distribuição dos pontos de coleta de amostras - Pará, Brasil ')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Adicionar legenda
ax.legend()

# Salvar o plot em formato PDF
plt.savefig('pontos_coleta.svg', bbox_inches='tight')

plt.show()
```
